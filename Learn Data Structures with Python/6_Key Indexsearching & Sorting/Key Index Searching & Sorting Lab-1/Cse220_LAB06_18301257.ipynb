```python
class Keyindex():
  
  def __init__(self,ab):
    self.min=self.minx(ab)
    self.max_1=self.maxy(ab)
    self.max=self.max_1-self.min

    self.ar=[0]*len(ab)
    for i in range(0,len(ab)):
      self.ar[i]=ab[i]-self.min
    
    self.aux=[0]*(self.max+1)

    for i in range(0,len(self.ar)):
      self.aux[self.ar[i]]+=1
    
  def minx(self,ab):
    min=0
    for i in ab:
      if i<=min:
        min=i
    return min
  def maxy(self,ab):
    max=0
    for j in ab:
      if j>max:
        max=j
    return max
    
  def search(self,v):
      for i in range(0,len(self.aux)):
        if self.aux[v-self.min]>0:
          return True
        else:
          return False
    
  def sort(self):
      for i in range(0,len(self.aux)):
        if self.aux[i]>0:
          for j in range(0,self.aux[i]):
            #update given array then print ab
            print(i+self.min,end=' ')



arr=[-2,-8,3,5,5]
h=Keyindex(arr)
print(h.search(3))
h.sort()
```

```python
vowels = ["A", "E", "I", "O", "U"]
consonants = ["B", "C", "D", "F", "G", "H", "J", "K", "L",
              "M", "N", "P", "Q", "R", "S", "T", "V", "W", "X", "Y", "Z"]


class Hashstring:
    def __init__(self, str):
        self.str = str

    def string_get(self):
        return self.str

    def string_key(self):
        val = 0
        n = 0
        for x in self.str:
            if x in consonants:
                val = val + 1
            if x not in consonants and x not in vowels:
                n += int(x)
        s_key = ((val * 24) + n)
        return s_key


class Hashtable:
    def __init__(self, size=9):
        self.size = size
        self.list = [0] * size

    def function(self, string_key):
        return string_key % self.size

    def add(self, s):
        val = s.string_key()
        position = self.function(val)
        for y in range(1, self.size):
            if self.list[position] == 0:
                self.list[position] = s.string_get()
                break
            else:
                position = (val+y) % self.size

    def print(self):
        for y in range(self.size):
            print(self.list[y])


hash_table = Hashtable()
list = ['ST1E89B8A32']
for x in list:
    v = Hashstring(x)
    hash_table.add(v)
hash_table.print()
```
